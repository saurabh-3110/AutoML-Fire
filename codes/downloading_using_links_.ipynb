```python
import os
import requests
from urllib.parse import unquote, urlparse, parse_qs

# Function to download a file from a given URL
def download_file(url, destination):
    response = requests.get(url)
    with open(destination, 'wb') as file:
        file.write(response.content)

# Read links from the text file
file_path = '2005-2023_download_link_files/2009_soil_moisture&aerodynamic_conductance.txt'  # Replace with the path to your text file
with open(file_path, 'r') as file:
    links = file.read().splitlines()

# Create a new folder for downloads
download_folder = '2009'
os.makedirs(download_folder, exist_ok=True)

# Download each file
for i, encoded_link in enumerate(links, 1):
    decoded_link = unquote(encoded_link)
    
    # Extract filename from the decoded link
    parsed_url = urlparse(decoded_link)
    filename = parse_qs(parsed_url.query).get('LABEL', [''])[0]
    
    # Check if the filename has a .nc4 extension
    if filename and filename.endswith('.nc4'):
        # Specify the destination file path in the new folder
        destination_file = os.path.join(download_folder, filename)
        
        download_file(decoded_link, destination_file)
        print(f'Downloaded: {destination_file}')
    else:
        print(f'Skipped: Invalid file type for link {i}')

print('Download completed!')
```
